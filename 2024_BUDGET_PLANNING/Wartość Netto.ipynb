```python
import matplotlib.pyplot as plt

# Dane wejciowe
aktywa = {
    "Meble wolnostojce": 400,
    "Sprzt elektroniczny (RTV)": 7850,
    "Inne rodki transportu (rower, skuter)": 800,
    "Biuteria": 2150,
    "Akcje, udziay w firmach": 27000,
    "Obligacje IKE": 24262,
    "Obligacje": 39428,
    "Lokaty w banku": 0,
    "Konto oszczdnociowe": 3200,
    "Rachunek bankowy": 14000,
    "Waluty obce": 200,
    "Gotwka": 50,
    "PPK": 0,
}

zobowiazania = {
    "Poyczka od rodziny": 0,
    "Ksigowa": 0,
    "Podatki": 1,
    "Studia": 0,
}

wartosc_netto = [77288, 83288, 97467, 119339]

daty = [
    "05/06/2024", "06/07/2024", "07/08/2024", "12/12/2024"
]

# Obliczenia sum
suma_aktywa = sum(aktywa.values())
suma_zobowiazania = sum(zobowiazania.values())
wartosc_netto_obliczona = suma_aktywa - suma_zobowiazania

# Wykres 1: Koowy aktyww i zobowiza
fig1, ax1 = plt.subplots(1, 2, figsize=(12, 6))
ax1[0].pie(aktywa.values(), labels=aktywa.keys(), autopct='%1.1f%%', startangle=140)
ax1[0].set_title('Rozkad Aktyww')
ax1[1].pie(zobowiazania.values(), labels=zobowiazania.keys(), autopct='%1.1f%%', startangle=140)
ax1[1].set_title('Rozkad Zobowiza')

# Wykres 2: Supkowy aktyww i zobowiza
fig2, ax2 = plt.subplots(figsize=(10, 6))
aktywa_keys = list(aktywa.keys())
aktywa_values = list(aktywa.values())
zobowiazania_keys = list(zobowiazania.keys())
zobowiazania_values = list(zobowiazania.values())
ax2.barh(aktywa_keys, aktywa_values, color='green', label='Aktywa')
ax2.barh(zobowiazania_keys, zobowiazania_values, color='red', label='Zobowizania')
ax2.set_xlabel('Warto (PLN)')
ax2.set_title('Porwnanie Aktyww i Zobowiza')
ax2.legend()

# Wykres 3: Warto netto w czasie
fig3, ax3 = plt.subplots(figsize=(10, 6))
ax3.plot(daty, wartosc_netto, marker='o', color='blue')
ax3.set_xlabel('Data')
ax3.set_ylabel('Warto netto (PLN)')
ax3.set_title('Zmiana Wartoci Netto w Czasie')
ax3.grid(True)

# Wywietlenie wszystkich wykresw
plt.tight_layout()
plt.show()

# Wypisanie sumy aktyww, zobowiza i wartoci netto
print(f"Suma aktyww: {suma_aktywa} z")
print(f"Suma zobowiza: {suma_zobowiazania} z")
print(f"Twoja warto netto wynosi: {wartosc_netto_obliczona} z")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Przykadowe dane o czasie spdzonym na zadaniach
data = {
    "Zadanie": ["Praca", "Nauka", "Relaks", "wiczenia", "Zakupy", "Praca In"],
    "Czas (godz.)": [6, 3, 2, 3, 1, 3]
}

# Tworzenie DataFrame z danych
df = pd.DataFrame(data)

# Funkcja do wywietlania wykresu koowego
def show_pie_chart(df):
    plt.figure(figsize=(8, 6))
    plt.pie(df["Czas (godz.)"], labels=df["Zadanie"], autopct='%1.1f%%', startangle=140)
    plt.title('Podzia czasu na zadania')
    plt.show()

# Funkcja do dodawania nowych danych
def add_task(df, task_name, hours):
    global data
    data["Zadanie"].append(task_name)
    data["Czas (godz.)"].append(hours)
    return pd.DataFrame(data)

# Dodaj nowe zadanie (przykad)
df = add_task(df, "Czytanie", 2)

# Wywietl wykres
show_pie_chart(df)
```
